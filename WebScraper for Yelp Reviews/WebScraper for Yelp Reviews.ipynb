I designed my scraping through using yelp and picking restaurants that had lots of good reviews and a high rating.
Then I inspected the page of a certain restraurant (starting with guacho) and I found the input for the tags to 
pull information on using beautiful soup. Using these inputs I got, I was able to get the information for reviews,
stars, dates, authors, and the usefulness factor. I ran into lots of issues trying to get the exact data I needed
as several times it kept pulling unessesary information, when someone updated their review, or from ads about the
ratings of other restaurants. To combat this I narrowed my search even more using extra commands like .find and .get
so I can only get the exact information I need.
To minimize my impact on yelp I spent a couple days testing and creating this so I would not just continuously
keep going through their pages. I also tests with just going through 2-3 pages instead of all 10-16 pages of reviews


```python
import bs4
import requests
import pandas
import numpy
import re
#schools = pandas.DataFrame(schools, columns=['ranking','school','conference'])
#schools
# numpycolumn_stack
#url = 'https://www.yelp.com/search?find_desc=Restaurants&find_loc=Blacksburg%2C%20VA'
header = {'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/12.0.3 Safari/605.1.15'}
url = 'https://www.yelp.com/biz/gaucho-brazilian-grille-blacksburg-2'
review = []
star = []
dates = []
authors = []
usefuls = []
name = []
for i in range(0,6): #6
    urlappend = '?start=' + str(i*20)
    url2 = url + urlappend
    page = requests.get(url2, headers=header)
    soup = bs4.BeautifulSoup(page.text, 'html5lib')
    
    reviews = soup.find_all('p', {'lang' : 'en'})
    
    stars = soup.find_all('div', {'class' : 'review review--with-sidebar'})
  
    date = soup.find_all('div', {'class' : 'review review--with-sidebar'})
    
    author = soup.find_all('a', {'class' : 'user-display-name js-analytics-click'})
    
    useful= soup.find_all('div', {'class' : 'review review--with-sidebar'})
    
    
    
    
    #make the lists
    for j in range(0, 16): #16
        review.append(reviews[j].text);
        
        star.append(stars[j].find('div', {'class' : re.compile('i-stars i-stars--regular-.* rating-large')}).find('img').get('alt'))
        
        dates.append(date[j].find('span', {'class' : 'rating-qualifier'}).text.split()[0])
        
        authors.append(author[j].text)
        usefuls.append(useful[j].find('div', { 'class' : 'rateReview voting-feedback'}).find('p', {'class' : 'voting-intro voting-prompt'}).text.strip())
        #dates.append(date[j].text.split()[0])
        name.append("Gaucho Brazilian Grille")
        
len(review)      
```
